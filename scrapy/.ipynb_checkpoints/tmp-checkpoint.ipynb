```python
def get_jisilu_cb(url):
    soup = get_soup(url)
    table= []
    for tr in soup.find_all('tr')[2:]:
        lst = []
        for td in tr.find_all("td"):
            lst.append(td.get_text())
        table.append(lst)
    return table
```

```python
tbl_lst = get_jisilu_cb(url)
```
